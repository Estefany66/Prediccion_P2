```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from datetime import datetime
```

```python
# Cargar datos desde un archivo CSV
data = pd.read_csv("/content/drive/MyDrive/Fecha3.csv")

# Asegúrate de tener una columna de fecha en formato adecuado (puedes necesitar ajustarla)
data['Fecha'] = pd.to_datetime(data['Fecha'])

# Ordena los datos por fecha si no están ordenados
data.sort_values(by='Fecha', inplace=True)

# Ajusta los nombres de columnas según tu estructura de datos
data
```

```python
# Crea características adicionales si es necesario (por ejemplo, estacionalidad)
data['Mes'] = data['Fecha'].dt.month
data['Año'] = data['Fecha'].dt.year

# Divide los datos en características (X) y ventas (y)
X = data[['Mes', 'Año']]  # Ajusta las características según tus datos
y = data['Venta']
```

```python
# Crea el modelo Random Forest para regresión
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)  # Puedes ajustar los hiperparámetros según sea necesario

# Entrena el modelo
rf_model.fit(X, y)
```

```python
# Genera una fecha futura (por ejemplo, 12 meses después de la última fecha en tus datos)
ultima_fecha = data['Fecha'].max()
fecha_futura = ultima_fecha + pd.DateOffset(months=12)
```

```python
# Crea un conjunto de características para la fecha futura
mes_futuro = fecha_futura.month
año_futuro = fecha_futura.year
features_futuro = [[mes_futuro, año_futuro]]

# Realiza la predicción
prediccion_futura = rf_model.predict(features_futuro)

# Imprime la predicción
print(f"Predicción de ventas para {fecha_futura}: {prediccion_futura[0]}")
```
